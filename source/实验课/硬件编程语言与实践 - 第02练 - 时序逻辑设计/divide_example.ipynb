```python
# FULL PRECISION

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

fp_a = open('DIVIDE_DAT_A_S0I8F0.dat', 'w')
fp_b = open('DIVIDE_DAT_B_S0I8F0.dat', 'w')
fp_c = open('DIVIDE_DAT_C_S0I8F8.dat', 'w')
for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE
            # dump
            fp_a.write("{:x}\n".format(a))
            fp_b.write("{:x}\n".format(b))
            fp_c.write("{:x}\n".format(int(c_target * DATA_SCALE)))
fp_a.close()
fp_b.close()
fp_c.close()
```
