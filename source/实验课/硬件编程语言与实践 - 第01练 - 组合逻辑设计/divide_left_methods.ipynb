```python
# USE REMAINDER NON-RESTORING METHOD (IN A "MOST" HARDWARE WAY)

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! use remainder non-restoring method (in a "most" hardware way) (LESS recommended)
            # !!! overly detailed modeling is more prone to introducing errors.
            NUMB_BIT   = DATA_WD * 2 + 1
            DATA_SHIFT = DATA_WD - 1
            DATA_MASK  = 2 ** DATA_SHIFT - 1
            t = a
            c_in_rtl = 0
            d_in_rtl = (t >> DATA_SHIFT) - b
            t = (d_in_rtl << DATA_SHIFT) | (t & DATA_MASK)
            for idxBit in range(NUMB_BIT):
                t = t << 1
                if t < 0:
                    d_in_rtl = (t >> DATA_SHIFT) + b
                    t = (d_in_rtl << DATA_SHIFT) | (t & DATA_MASK)
                    c_in_rtl = c_in_rtl << 1
                else:
                    d_in_rtl = (t >> DATA_SHIFT) - b
                    t = (d_in_rtl << DATA_SHIFT) | (t & DATA_MASK)
                    c_in_rtl = (c_in_rtl << 1) + 1
            c_in_rtl = (c_in_rtl + 1) >> 1
            c = c_in_rtl / DATA_SCALE
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
```

```python
# BINARY SEARCH

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! binary search
            DATA_SHIFT_R = DATA_WD + 16
            DATA_SCALE_R = 2 ** DATA_SHIFT_R
            DATA_SCALE_C = 2 ** DATA_WD
            r = 0
            for idxBit in range(1 + DATA_SHIFT_R + 1):
                r = r + (2 ** (- idxBit))
                if r * b > 1:
                    r = r - (2 ** (- idxBit))
            r = floor(r * DATA_SCALE_R + 0.5) / DATA_SCALE_R
            c = a * r
            c = floor(c * DATA_SCALE_C + 0.5) / DATA_SCALE_C
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
```

```python
# USE NEWTON-RAPHSON ITERAION (WITH LIMITED PRECISION)

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! use Newton-Raphson iteraion (with limited precision)
            NUMB_ITER    = 12
            DATA_SHIFT_T = DATA_WD + 16
            DATA_SCALE_T = 2 ** DATA_SHIFT_T
            DATA_SCALE_C = 2 ** DATA_WD
            t = 1 / (2 ** DATA_WD)
            for idxItr in range(NUMB_ITER):
                j = b * t
                j = floor(j * DATA_SCALE_T + 0.5) / DATA_SCALE_T
                k = 2 - j
                t = t * k
                t = floor(t * DATA_SCALE_T + 0.5) / DATA_SCALE_T
            c = floor(a * t * (2 ** DATA_WD) + 0.5) / DATA_SCALE_C
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
```

```python
# USE NEWTON-RAPHSON ITERAION (WITH LIMITED PRECISION) (WITH STATISTICS)

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

j_max = 0
k_max = 0
t_max = 0
for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! use Newton-Raphson iteraion (with limited precision) (with statistics)
            NUMB_ITER    = 12
            DATA_SHIFT_T = DATA_WD + 16
            DATA_SCALE_T = 2 ** DATA_SHIFT_T
            DATA_SCALE_C = 2 ** DATA_WD
            t = 1 / (2 ** DATA_WD)
            for idxItr in range(NUMB_ITER):
                #--- J ---
                # perform core calculation
                j = b * t
                # perform rounding
                j = floor(j * DATA_SCALE_T + 0.5) / DATA_SCALE_T
                # perform statistics
                if j_max < j:
                    j_max = j

                #--- K ---
                # perform core calculation
                k = 2 - j
                # perform rounding
                # not necessary
                # perform statistics
                if k_max < k:
                    k_max = k

                #--- T ---
                # perform core calculation
                t = t * k
                # perform rounding
                # not necessary
                t = floor(t * DATA_SCALE_T + 0.5) / DATA_SCALE_T
                # perform statistics
                if t_max < t:
                    t_max = t

            c = floor(a * t * (2 ** DATA_WD) + 0.5) / DATA_SCALE_C
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
print(f"max of j, k, t is: {j_max}, {k_max}, {t_max}")
```
