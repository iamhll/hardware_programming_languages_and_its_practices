```python
# FULL PRECISION

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

fp_a = open('DIVIDE_DAT_A_S0I8F0.dat', 'w')
fp_b = open('DIVIDE_DAT_B_S0I8F0.dat', 'w')
fp_c = open('DIVIDE_DAT_C_S0I8F8.dat', 'w')
for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE
            # dump
            fp_a.write("{:x}\n".format(a))
            fp_b.write("{:x}\n".format(b))
            fp_c.write("{:x}\n".format(int(c_target * DATA_SCALE)))
fp_a.close()
fp_b.close()
fp_c.close()
```

```python
# DIRECTLY USE "/" (IN A MORE HARDWARE WAY)

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! directly use "/" (in a more hardware way) (RECOMMENDED)
            # !!! it not only reflects the characteristic of the "/" operator, but also clearly expresses the meaning of both c_in_rtl and c.
            DATA_SHIFT = DATA_WD + 1
            c_in_rtl = (a << DATA_SHIFT) // b
            c_in_rtl = (c_in_rtl + 1) >> 1
            c = c_in_rtl / DATA_SCALE
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
```

```python
# USE REMAINDER RESTORING METHOD (FIX B SHIFT T) (IN A "MOST" HARDWARE WAY)

# clear all variables
%reset -f

import random
from math import floor

DATA_WD = 8

for a in range(2 ** DATA_WD):
    for b in range(2 ** DATA_WD):
        if b != 0:
            # !!! full precision
            DATA_SCALE = 2 ** DATA_WD
            c_target = a / b
            c_target = floor(c_target * DATA_SCALE + 0.5) / DATA_SCALE

            # !!! use remainder restoring method (fix b shift t) (in a "most" hardware way) (LESS recommended)
            # !!! overly detailed modeling is more prone to introducing errors.
            NUMB_BIT   = DATA_WD * 2 + 1
            DATA_SHIFT = DATA_WD - 1
            DATA_MASK  = 2 ** DATA_SHIFT - 1
            t = a
            c_in_rtl = 0
            for idxBit in range(NUMB_BIT):
                d_in_rtl = (t >> DATA_SHIFT) - b
                t = (d_in_rtl << DATA_SHIFT) | (t & DATA_MASK)
                if t < 0:
                    d_in_rtl = (t >> DATA_SHIFT) + b
                    t = (d_in_rtl << DATA_SHIFT) | (t & DATA_MASK)
                    c_in_rtl = c_in_rtl << 1
                else:
                    c_in_rtl = (c_in_rtl << 1) + 1
                t = t << 1
            c_in_rtl = (c_in_rtl + 1) >> 1
            c = c_in_rtl / DATA_SCALE
            if (c != c_target):
                print("ERROR!")
                print(f"{a} / {b} = {c_target}")
                print(f"{a} / {b} = {c}")
                break
    if (c != c_target):
        break
if (c == c_target):
    print("TEST PASSED!")
```
